```python
cut_off = 50
num_programs = 22
guides_per_program = {}

for i in range(1, num_programs + 1):
    counts = Counter()
    col_name = f"{i}" 
    
    top_scores = usage.nlargest(cut_off, col_name)
    ids = top_scores["Unnamed: 0"].to_list()
    
    for j in ids:
        if j not in gdata.obs.index:
            continue
            
        guide_str = gdata.obs.loc[j, 'called_guides']
        
        if pd.isna(guide_str) or guide_str == "":
            continue
            
        g_arr = [g.strip() for g in str(guide_str).split(',')]
        counts.update(g_arr) 
        
    guides_per_program[i] = dict(counts)
```

```python
guide_program_df = pd.DataFrame(guides_per_program)
top_n = 3
program_top_guides = {}


for i in range(1, num_programs+1):
    top_scores = guide_program_df.nlargest(top_n, i)
    guides = top_scores.index.to_list()
    program_top_guides[i]=guides


top_guide_program = pd.DataFrame(program_top_guides)
```
